```python
import tweepy

consumerkey = ""
consumersecret = ""
accesstoken = ""
accesstokensecret = ""

auth = tweepy.OAuthHandler(consumerkey, consumersecret)
auth.set_access_token(accesstoken, accesstokensecret)

api = tweepy.API(auth)

public_tweets = api.home_timeline()
for tweet in public_tweets:
    print(tweet.text)
```
