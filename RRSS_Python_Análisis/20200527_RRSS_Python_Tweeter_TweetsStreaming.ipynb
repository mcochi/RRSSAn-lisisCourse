```python
import pandas as pd
```

```python
datos = []
for tweet in tweets:
    d = {'id':tweet.id,
        'lang': tweet.lang,
        'text': tweet.text,
        'user_id':tweet.user.name,
        'screen_name':tweet.user.screen_name,
        'name':tweet.user.screen_name,
        'status_count':tweet.user.statuses_count,
        'retweeet_count':tweet.retweet_count}
    datos.append(d)
    
df = pd.DataFrame(datos)
```
